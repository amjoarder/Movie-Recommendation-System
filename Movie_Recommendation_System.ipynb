```python
import pandas as pd
import numpy as np
```

```python
movie = pd.read_csv('dataset.csv')
```

```python
movie.head()
```

```python
movie.columns
```

```python
movie.info()
```

```python
movie['tags'] = movie['genre'] + movie['overview']
```

```python
movie.head()
```

```python
new_df = movie[['id','title','tags']]
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vec = cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
```

```python
vec
```

```python
vec.shape
```

#**Find Cosine Similarity**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vec)
```

```python
sim
```

*Sort Movies according to Similarity*

```python
new_df[new_df['title'] == 'The Lego Movie']
```

**Movies Simiar to "The Lego Movie" in Descending Order**

```python
dist = sorted(list(enumerate(sim[1632])),reverse=True,key=lambda vec:vec[1])
```

```python
dist
```

```python
for i in dist[0:10]:
  print(new_df.iloc[i[0]].title)
```

#Creating a Function

```python
def recommend(movie):
  index = new_df[new_df['title'] == movie].index[0]
  distance = sorted(list(enumerate(sim[index])),reverse=True,key=lambda vec:vec[1])
  for i in distance[0:10]:
    print(new_df.iloc[i[0]].title)
```

**Test With a Random Movie Name**

```python
recommend('The Shawshank Redemption')
```
